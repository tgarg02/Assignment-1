## 1) Implement 10 combinations of Data operators ( +, -, *, /, //, %) between 2 simple arrays.
---

```python
# Two arrays a and b
a = [2,3,4,8]
b = [5,2,7,8]
n = min(len(a),len(b))
```

```python
#Multiplication
print("Multiplicatiion of two arrays : ")
for i in range(n):
  print(a[i] * b[i])
```

```python
#Addition
print("Addtion of two arrays : ")
for i in range(n):
  print(a[i] + b[i])
```

```python
#Subtraction
print("Subtraction of two arrays : ")
for i in range(n):
  print(a[i] - b[i])
```

```python
#Division
print("Division of two arrays : ")
for i in range(n):
  print(a[i] / b[i])
```

```python
#Multiplication and Division
print("Division by 2 to the multiplication of two arrays  : ")
for i in range(n):
  print((a[i] * b[i])/2)
```

```python
#Floor Division
print("Floor Divison of two arrays : ")
for i in range(n):
  print(a[i] // b[i])
```

```python
#Modulus
print("Remainder of the division of two arrays : ")
for i in range(n):
  print(a[i] % b[i])
```

```python
#Power
print("Power of two arrays : ")
for i in range(n):
  print(a[i] ** b[i])
```

```python
#Multiplication and Addtion
print("Multiplication of the sum of two arrays by two : ")
for i in range(n):
  print((a[i] + b[i])*2)
```

```python
#Modulus and Subtraction
print("Subtration of 2 from the modulus of two arrays  : ")
for i in range(n):
  print((a[i] % b[i])-2)
```

## 2) Implement 2 examples each of " For " loop and " While " loop
---

```python
# For Loop

# Example 1
print("This is Example 1 : ")
arr = [20,30,44,7,5]
for i in arr:
  print(i)

# Example 2
print("This is Example 2 : ")
for i in range(0,5):
  print(i)
```

```python
# While Loop

#Example 1
print("This is Example 1 : ")
i = 0
while i < 5:
  print(i)
  i+=1

#Example 2
print("This is Example 2 : ")
i = 2
while i <= 10:
  print(i)
  i+=2
```
